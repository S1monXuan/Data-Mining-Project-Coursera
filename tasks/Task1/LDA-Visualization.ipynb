```python
sample = pd.read_csv('./results/LDA-Sample.csv', sep=' ', header=None, index_col=0)
sample.index.names = ['Topic']
pos = pd.read_csv('./results/LDA-Positive.csv', sep=' ', header=None, index_col=0)
pos.index.names = ['Topic']
neg = pd.read_csv('./results/LDA-Negative.csv', sep=' ', header=None, index_col=0)
neg.index.names = ['Topic']
```

# Create Radial Tree Layout for All Sample

```python
def node_create(data):
    node_data = [{"label":"root", "shape":"rect"}]
    link_data = []
    cnt = 0
    for idx, row in data.iterrows():
#         print(row)
        node_data.append({"label": idx, "shape":"rect"})
        idxN = cnt + 1
        link_data.append({"source":0, "target":idxN})
        times = 1
        for item in row:
            node_data.append({"label": item, "shape":"rect"})
            link_data.append({"source":idxN, "target":idxN + times})
            times+=1
        cnt += (1 + len(row))
    return node_data, link_data
# add Topic
```

# create graph

## create sample graph

```python
node_data, link_data = node_create(sample)
```

```python
print(node_data)
```

```python
print(link_data)
```

```python
fig, ax= plt.subplots(figsize=(100, 100))

sample_graph = bqplot.Graph(node_data = node_data,
                           link_data = link_data,
                            link_distance = 60,
                            colors = 'gray'
                           )
sample_graph.directed = False
sample_graph.link_type = 'line'
sample_graph.charge = -200

fig = bqplot.Figure(marks=[sample_graph])
fig.layout.min_width='1000px'
fig.layout.min_height = '1000px'
fig
```

# Task1-2 based on sentiment

```python
def node_sent_create(data, startNum: int, node_data, link_data):
#     node_data = [{"label":"root", "shape":"rect"}]
#     link_data = []
    cnt = startNum
    for idx, row in data.iterrows():
#         print(row)
        node_data.append({"label": idx, "shape":"rect"})
        idxN = cnt + 1
        link_data.append({"source":0, "target":idxN})
        times = 1
        for item in row:
            node_data.append({"label": item, "shape":"rect"})
            link_data.append({"source":idxN, "target":idxN + times})
            times+=1
        cnt += (1 + len(row))
    return node_data, link_data
```
