# Task 4: Mining Popular Dishes

```python
data
```

```python
# count occurences of all selected name and its number 
def analysis_cuisine(cuisine_name, data):
    count = 0
    rating = 0
    for idx, vals in data.iterrows():
        if cuisine_name in vals['text']:
            count+=1
            rating+=vals['stars']
    if count > 0:
        rating /= count 
    else:
        rating = 0
    return count, rating
```

```python
print(analysis_cuisine('rice', data))
```

```python
print(analysis_cuisine('adsfasdf', data))
```

```python
def get_cuisine_dict(cuisine_list, data):
    res = []
    for cuisine in cuisine_list[0]:
        cur = []
        cur.append(cuisine)
        tmp = analysis_cuisine(cuisine, data)
        if tmp[0] == 0:
            continue
        else:
            cur.append(tmp[0])
            cur.append(tmp[1])
            res.append(cur)
    return res
```

```python
cuisine_list = get_cuisine_dict(cuisine_data.head(10), data)
cuisine_list
```

```python
cuisine_list = get_cuisine_dict(cuisine_data, data)
cuisine_list
```

# Create pandas for visualization

```python
task4_data = pd.DataFrame(cuisine_list, columns=['cuisine', 'occurences', 'rating'])
task4_data = task4_data.sort_values(by=['occurences', 'rating'], axis = 0, ascending=[True, False], inplace=False, ignore_index = True, key = None)
```

# Task 5, Restaurant Recommendation

```python
cuisine = 'fried rice'

# get restaurant list
rest_lists = data['name'].unique()
rest_lists
```

```python
# count occurance and ratings for each restaurant based on particular cuisine
def get_rest_list(rest_list, cuisine, data):
    res = []
    for rest in rest_list:
        cur = []
        stars = 0
        count = 0
        data_tmp = data[data['name'] == rest]
        for idx, review in data_tmp.iterrows():
            if cuisine in review['text']:
                stars += review['stars']
                count += 1
        if count == 0: continue
        cur.append(rest)
        cur.append(count)
        cur.append(stars/count)
        res.append(cur)
    return res  
            
```

```python
task5_list = get_rest_list(rest_lists, cuisine, data)
task5_list
```

# Visualization for Task 5

```python
task5_data = pd.DataFrame(task5_list, columns=['restaurant', 'occurences', 'rating'])
task5_data = task5_data.sort_values(by=['occurences', 'rating'], axis = 0, ascending=[True, False], inplace=False, ignore_index = True, key = None)
task5_data
```

```python
task5 = alt.Chart(task5_data.tail(100)).mark_bar().encode(
    alt.X('restaurant:O', axis = alt.Axis(labelAngle=45), sort=None),
    alt.Y('max(occurences):Q', title="Occurences"),
    alt.Color('rating:Q', scale=alt.Scale(scheme='magma'))
).properties(
    width=1000,
    title='Task_5_Restaurant_Recommendation_for_fried_rice'
)


task5
```
